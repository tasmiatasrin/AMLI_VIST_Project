```python
# grab story dictionary

import os

import json

# open JSON file
# f = open('train_split1_stories.json')
d = open('tsplit1_dictionary.json')

# returns JSON object as a dictionary
# tsplit1_data = json.load(f)
tsplit_dict = json.load(d)

# close file
# f.close
d.close
```

```python
# grab the keys
tsplit_keys = []
for key in tsplit_dict.keys():
    tsplit_keys.append(key)

print(len(tsplit_keys))
```

```python
# IMPORTANT
# remove stories with missing images
import os
import pickle as pkl

missing_stories = []

for key in tsplit_keys: # grab the story_ids
    for image in tsplit_dict[key]['images']: # image list of each story_id
        
        if image + '.pkl' not in os.listdir(pickle_dir): # check which stories have missing pkl files
            if key not in missing_stories:
                missing_stories.append(key)
                print(len(missing_stories))
            
missing_stories

for story in missing_stories:
    tsplit_keys.remove(story)
    
print(f'Finished! Length of keys: {len(tsplit_keys)}')
```

```python
import os
import pickle as pkl

os.chdir(pickle_dir) # change to pickle folder

features = []
five_features = []
new_len = 0 # used for printing

missing_stories = []

for key in tsplit_keys: # grab the story_ids
    for image in tsplit_dict[key]['images']: # image list of each story_id
        
        if image + '.pkl' in os.listdir(pickle_dir): # check if in pickle_dir
            pkl_file = open(image + '.pkl', 'rb') # open pickle file
            one_feature = pkl.load(pkl_file) # grab feature
                
            features.append(one_feature) # store features

            if len(features) == 5:
                five_features.append(features) # store the set of 5 image features
                features = []

            # loading print
            old_len = len(five_features)
            if len(five_features)%100 == 0 and old_len != new_len:
                new_len = old_len
                print(len(five_features))
                
print(f'Finished! Length: {len(five_features)}')

os.chdir(home_dir) # change back to home folder

! pwd
```
